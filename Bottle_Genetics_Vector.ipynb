```
import numpy as np
import matplotlib.pyplot as plt
import random
import copy
```

Code to give multiple points between 2 points which is used to draw line between them.

```
def draw_line(p1,p2,canvas_size, canvas):
  
  x_diff=abs(p1[0]-p2[0])  
  y_diff=abs(p1[1]-p2[1]) 
  x_division=x_diff/(canvas_size-1)
  y_division=y_diff/(canvas_size -1)
  x=p1[0]
  y=p1[1]
  
  while round(x)!=p2[0] or round(y)!=p2[1]: 
    if p1[0]<p2[0] :
      x=x+x_division 
    elif p1[0]>p2[0]: 
      x=x-x_division 
    if p1[1]<p2[1]: 
      y=y+y_division 
    elif p1[1]>p2[1] :
      y=y-y_division 
    X, Y=round(x), round(y)
    canvas[X][Y]=1
  return canvas



  
```

## Drawing an image using 8 points

```
def img(p, canvas_size):
  canvas=np.array([[0]*canvas_size for i in range(canvas_size)])
  for i in range(len(p)-1):
    canvas=(draw_line(p[i], p[i+1], canvas_size,canvas))
    if i+1==len(p)-1:
      canvas=np.array(draw_line(p[i+1], p[0], canvas_size, canvas))
  
  plt.imshow(canvas, cmap='gray',origin='lower')
  plt.show()
```

### Generate Population:
* Generate random population(8 points) of given population size.

### Fitness function
* Compares the population with the original image and calculates the fitness score.

### Cross breed:
* This function will take two members from population as input and generate two new offsprings using combination of both members.

### Mutate
* To maintain genetic diversity from one generation of a population of genetic algorithm to the next we are altering one point.

```
def generate_population(population_size,points,canvas_size):
  population=[]
  for pop in range(population_size):
    p1=[0]
    p=[[0,0]]
    for i in range(points-1):
      np.array(p.append([random.randint(0,canvas_size-1),random.randint(0,canvas_size-1) ]))
    p1=p1+(p)
    population.append(p1)
  return np.array(population)

def fitness_func(member, points):
  p=[[0,0],[0,10],[10,10],[15,7],[17,7],[17,4],[15,4],[10,0]]
  p=list(map(lambda p1: [p1[0]*5 , p1[1]*5], p))
  fit=0
  for i in range(2,points+1):
    fit=fit+(canvas_size-abs(member[i][0]-p[i-1][0])+canvas_size-abs(member[i][1]-p[i-1][1]))
  member[0]=fit
  return(member)
   
def cross_breed(member1, member2):
  i=random.randint(2,len(member1)-1)
  m1=np.append(copy.deepcopy(member1[:][:i]), copy.deepcopy(member2[:][i:]))
  m2=np.append(copy.deepcopy(member2[:][:i]), copy.deepcopy(member1[:][i:]))
  return  m1,m2 

def mutate(member, points, canvas_size):
  index=random.randint(2,points)
  member[index][0]=random.randint(0,canvas_size-1)
  member[index][1]=random.randint(0,canvas_size-1)
  return member
```

```

population_size=100
generations=800
points=8
kill_rate=3
population=generate_population(population_size,points,canvas_size)
#print("initial population", population)


maxscore, minscore, distinctmaxfit=[],[],[]
for i in range(generations):
  j=0
  population=np.array(list(map(lambda ps: fitness_func(ps, points), population)))
  population=np.array(sorted(population, key=lambda a: a[0]))
  #print("Sorted population", population)
  maxscore.append(population[-1][0])
  minscore.append(population[0][0])
  print("Generation, maxscore, minscore", i, population[-1][0], population[0][0])
  #if i%50==0:
  #  img(population[-1][1:],canvas_size)
  if population[-1][0]>1200 and  population[-1][0] not in distinctmaxfit:
    distinctmaxfit.append(population[-1][0])
    print("Fitness score:", population[-1][0])
    img(population[-1][1:],canvas_size)
  population=population[population_size//kill_rate:]
  #print("Remaining population", population)
  popl=len(population)
  while len(population)<population_size:
    firstbreed=random.randint(0,popl-1)
    secondbreed= random.choice(list(filter(lambda ele: ele != firstbreed, range(0,popl)))) 
    population=np.append(population,cross_breed(population[firstbreed], population[secondbreed]),0)
  #print("Cross Bread population", population)
  for i, member in enumerate(population):
    if (random.random()<0.2):
      population[i]=mutate(member, points, canvas_size)
  
  
print(maxscore)
```

## Graph of Generation/ Min_Max_Score

```
plt.plot( range(generations),maxscore ,'b-', label="Maxscore")
plt.title("Generation/Min_Max_Score")
plt.plot( range(generations),minscore, 'r-',label="Minscore")
plt.xlabel("Generations")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
